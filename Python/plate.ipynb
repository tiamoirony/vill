```python
class Error101(Exception):
    def __str__(self):
        return ('번호판 자리가 맞지 않습니다.')


class Error202(Exception):
    def __str__(self):
        return ('번호판 끝에서 네자리가 맞지 않거나 숫자가 아닙니다.')

class Error303(Exception):
    def __str__(self):
        return ('번호판 끝에서 다섯자리 맞지 않거나 문자가 아닙니다.')
    
class Error404(Exception):
    def __str__(self):
        return ('한글로된 지역명만 올수 있습니다.')

class Error505(Exception):
    def __str__(self):
        return ('숫자여야합니다.')
    
class Error606(Exception):
    def __str__(self):
        return ('해당 자릿수가 맞지 않거나 숫자가 아닙니다.')



class Start():

    def __init__(self, plate):

        if len(plate) not in [7, 8, 9]:
            raise(Error101)

        if not (plate[-4:].isdigit()) :
            raise(Error202)
    
        if not (plate[-5]).isalpha():
            raise (Error303)


    def check(self, plate):

        if len(plate) == 9:
            if not (plate[:2]).isalpha():
                raise(Error404)
            if not(plate[2:4].isdigit()):
                raise(Error505)

        if len(plate) == 8:
            if not (plate[:3].isdigit()):
                raise(Error606)

        if len(plate) == 7:
            if not (plate[:2].isdigit()):
                raise(Error606)


try:
    a = Start('12가1234')
    a.check()
except (Error101, Error202, Error303, Error404, Error505) as e:
    print(e)
```
